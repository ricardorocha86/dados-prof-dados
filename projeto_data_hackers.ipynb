```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def analisar_salario(variavel):
    # Calcular estatísticas descritivas
    descritivas = data_new.groupby(variavel)['Faixa salarial'].describe()

    # Criar o boxplot
    plt.figure(figsize=(10, 6))  # Tamanho da figura (opcional)
    plot = sns.boxplot(x=variavel, y='Faixa salarial', data=data_new, showmeans = True, # Médias nas caixas: showmeans
                       flierprops=dict(marker='o', markersize=6, markerfacecolor='blue',
                                       markeredgecolor='black', alpha=0.05)) # Transparência do outlier: alpha

    # Ajustar o tamanho da fonte
    plot.set_xlabel(variavel, fontsize=10)  # Tamanho do rótulo do eixo X
    plot.set_ylabel('Faixa salarial', fontsize=10)  # Tamanho do rótulo do eixo Y
    plt.title('Boxplot da Faixa Salarial por ' + variavel, fontsize=12)  # Tamanho do título

    plt.xticks(fontsize=8)  # Tamanho da fonte para os rótulos do eixo X
    plt.yticks(fontsize=8)  # Tamanho da fonte para os rótulos do eixo Y

    # Adicionando as linhas horizontais em intervalos de 5 mil
    for i in range(1, 9):  # Adiciona linhas até 25 mil (1*5mil até 5*5mil)
        plt.axhline(y=i * 5000, color='gray', linestyle='--', linewidth=1)  # Linha horizontal a cada 5 mil

    # Fazendo o intervalo de confiança (alpha de 0.05)
    # Calcular a média e o erro padrão
    nomes = data_new.groupby(variavel)['Faixa salarial'].mean().index
    media = data_new.groupby(variavel)['Faixa salarial'].mean().to_list()
    desvio_padrao = data_new.groupby(variavel)['Faixa salarial'].std().to_list()

    n = len(data_new['Cargo'])
    # Calcular o intervalo de confiança
    intervalo = []
    for i in range(len(media)):
        ic_inf = media[i] - 1.96 * desvio_padrao[i]/np.sqrt(n)
        ic_sup = media[i] + 1.96 * desvio_padrao[i]/np.sqrt(n)
        intervalo.append((ic_inf, ic_sup))
        print(nomes[i], f'- IC Inf: {round(ic_inf, 2)}', f'IC Sup: {round(ic_sup, 2)}', '\n')


    plt.show()  # Mostrar o gráfico
    return descritivas, intervalo
```

## Criando nova função

Incluí funcionalidades na função analisar_salario:

- Salvar o plot em png para ser usado posteriormente no streamlit
- Estruturar o markdown para exibição dos intervalos de confiança
- Estruturar o markdown para exibição do sumário descritivo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def analisar_salario(variavel):
    # Calcular estatísticas descritivas ----------------------------------------
    descritivas = data_new.groupby(variavel)['Faixa salarial'].describe()

    # Criando o boxplot --------------------------------------------------------
    plt.figure(figsize=(10, 6))  # Tamanho da figura (opcional)
    plot = sns.boxplot(x=variavel, y='Faixa salarial', data=data_new, showmeans = True, # Médias nas caixas: showmeans
                       flierprops=dict(marker='o', markersize=6, markerfacecolor='blue',
                                       markeredgecolor='black', alpha=0.05)) # Transparência do outlier: alpha

    # Ajustar o tamanho da fonte
    plot.set_xlabel(variavel, fontsize=10)  # Tamanho do rótulo do eixo X
    plot.set_ylabel('Faixa salarial', fontsize=10)  # Tamanho do rótulo do eixo Y
    plt.title('Boxplot da Faixa Salarial por ' + variavel, fontsize=12)  # Tamanho do título

    plt.xticks(fontsize=8)  # Tamanho da fonte para os rótulos do eixo X
    plt.yticks(fontsize=8)  # Tamanho da fonte para os rótulos do eixo Y

    # Adicionando as linhas horizontais em intervalos de 5 mil
    for i in range(1, 9):  # Adiciona linhas até 25 mil (1*5mil até 5*5mil)
        plt.axhline(y=i * 5000, color='gray', linestyle='--', linewidth=1)  # Linha horizontal a cada 5 mil

    plt.savefig('grafico.png', dpi = 300, bbox_inches = 'tight') # Salvando o gráfico em png
    plt.close()

    # Fazendo o intervalo de confiança (alpha de 0.05) -------------------------
    # Calcular a média e o erro padrão
    nomes = data_new.groupby(variavel)['Faixa salarial'].mean().index
    media = data_new.groupby(variavel)['Faixa salarial'].mean().to_list()
    desvio_padrao = data_new.groupby(variavel)['Faixa salarial'].std().to_list()

    n = len(data_new['Cargo'])
    # Calcular o intervalo de confiança
    intervalo_inferior = []
    intervalo_superior = []
    for i in range(len(media)):
        ic_inf = media[i] - 1.96 * desvio_padrao[i]/np.sqrt(n)
        ic_sup = media[i] + 1.96 * desvio_padrao[i]/np.sqrt(n)
        intervalo_inferior.append(ic_inf)
        intervalo_superior.append(ic_sup)


    texto_markdown = f'''## Análise descritiva do Salário por {variavel} no mercado de dados no Brasil, 2023

### 📊 Sumário descritivo

{descritivas.to_markdown()}

### 📈 Visualização gráfica

![Salario por {variavel}]('grafico.png')

### 📏 Intervalo de confiança com nível de significância 0.05 para média de cada categoria'''
    for j in range(len(nomes)):
      texto_markdown += f'''

- {nomes[j]}:

Limitante Inferior - {round(intervalo_inferior[j], 2)}

Limitante Superior - {round(intervalo_superior[j], 2)}

'''

    return texto_markdown
```
